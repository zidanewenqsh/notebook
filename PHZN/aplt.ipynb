```python
def multishow(*args):
    plt.figure(figsize=(10, 6))
    size = len(args)
    n = math.ceil(size/2)
    for i in range(size):
        plt.subplot(n, 2, i+1)
        if args[i].ndim == 3:
            plt.imshow(args[i][...,::-1])
        else:
            plt.imshow(args[i])
    plt.show()
```
