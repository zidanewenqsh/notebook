## 旋转曲面 

```python
import numpy as np
import matplotlib.pyplot as plt

# 创建旋转曲线的参数
t = np.linspace(0, 2 * np.pi, 100)
x = 2 * np.cos(t)
y = 2 * np.sin(t)

# 创建旋转曲面的参数
theta = np.linspace(0, 2 * np.pi, 100)
z = np.outer(np.ones_like(theta), np.linspace(0, 1, 100))

# 绘制旋转曲面
X = x[:, np.newaxis] * np.cos(theta)
Y = x[:, np.newaxis] * np.sin(theta)
Z = z * np.ones_like(x[:, np.newaxis])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Rotating Surface')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(precision=2)
def hyperbola(x, a, b):
    idx = (x**2/a**2-1) >= 0
    y = np.zeros_like(x)
    y[idx] = b*np.sqrt(x[idx]**2/a**2 - 1)
    return x[idx], y[idx]
a = 1
b = 1
x = np.linspace(a, 4, 1000)
x, y = hyperbola(x, a, b)
# print(x, y)
x1, y1 = -x, -y
plt.plot(x, y)
plt.plot(x, y1)
plt.plot(x1, y)
plt.plot(x1, y1)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
np.set_printoptions(precision=2,suppress=True)
a = 1
b = 2
def discriminant(x, y):
    return -a**2*b**2 - a**2*y**2 + b**2*x**2
num = 1000
x_ = np.linspace(0, 3, num)
# x1_ = -np.linspace(1, 2, num)
x1_ = -x_
y_ = np.linspace(-5, 5, num)
x1, y1 = np.meshgrid(x_, y_)
x2, y2 = np.meshgrid(x1_, y_)
t1 = discriminant(x1, y1)
t2 = discriminant(x2, y2)
id1 = t1 >=0
id2 = t2 >=0

x1[~id1] = 1
x2[~id2] = -1
y1[~id1] = 0
y2[~id2] = 0
z1 = np.zeros_like(x1)
z2 = np.zeros_like(x2)
z1[id1] = np.sqrt(discriminant(x1[id1], y1[id1]))/b
z2[id2] = np.sqrt(discriminant(x2[id2], y2[id2]))/b
z3 = -z1
z4 = -z2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x1, y1, z1, color='b')
ax.plot_surface(x2, y2, z2, color='r')
ax.plot_surface(x1, y1, z3, color='g')
ax.plot_surface(x2, y2, z4, color='cyan')

ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 定义圆柱面的参数
radius = 1  # 圆柱面的半径
height = 5  # 圆柱面的高度

# 创建数据
theta = np.linspace(0, 2*np.pi, 100)  # 绕圆柱面轴线的角度
z = np.linspace(0, height, 100)      # 沿圆柱面轴线的高度
Theta, Z = np.meshgrid(theta, z)
Z.shape
# 计算圆柱面的坐标
X = radius * np.cos(Theta)
Y = radius * np.sin(Theta)
X.shape
Y.shape
# 绘制圆柱面
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.7)

# 添加标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Cylinder')

# 显示图形
plt.show()
```
